### • Crie gráficos para visualizar as diferenças entre as espécies de pinguins com base nas características físicas. Use gráficos de dispersão, boxplot e histogramas para mostrar as distribuições e comparações entre as variáveis.

```python
plt.figure(figsize=(10, 6))
sn.scatterplot(data=penguinsExtended_df, x="bill_length_mm", y="bill_depth_mm", size="body_mass_g", hue="species")
plt.xlabel("Comprimento do bico")
plt.ylabel("Profundidade do bico")
plt.show()
```

```python
penguinsExtended_df["species"].value_counts()
```

```python
adelie_df = penguinsExtended_df[penguinsExtended_df["species"]=="Adelie"]
gento_df = penguinsExtended_df[penguinsExtended_df["species"]=="Gentoo"]
chinstrap_df = penguinsExtended_df[penguinsExtended_df["species"]=="Chinstrap"]
```

```python
sn.set_style("whitegrid")
plt.boxplot([adelie_df["bill_length_mm"],gento_df["bill_length_mm"], chinstrap_df["bill_length_mm"]],labels=["Adelie","Gentoo","Chinstrap"])
plt.ylabel("Comprimento do bico")
plt.show()
sn.set_style("whitegrid")
plt.boxplot([adelie_df["bill_depth_mm"],gento_df["bill_depth_mm"], chinstrap_df["bill_depth_mm"]],labels=["Adelie","Gentoo","Chinstrap"])
plt.ylabel("Profundidade do bico")
plt.show()
sn.set_style("whitegrid")
plt.boxplot([adelie_df["body_mass_g"],gento_df["body_mass_g"], chinstrap_df["body_mass_g"]],labels=["Adelie","Gentoo","Chinstrap"])
plt.ylabel("Massa corporal")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.title("Caracteristicas Penguins Adelie")
plt.hist([adelie_df["bill_length_mm"],adelie_df["bill_length_mm"]], bins=50, label=["Comprimento do bico","Profundidade do bico"])
plt.legend(loc="upper right")
plt.show()
plt.figure(figsize=(8,5))
plt.title("Massa Corporal Penguins Adelie")
plt.hist([adelie_df["body_mass_g"]], bins=50, label=["Massa Corporal"],color="green")
plt.legend(loc="upper right")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.title("Caracteristicas Penguins Gentoo")
plt.hist([gento_df["bill_length_mm"],gento_df["bill_length_mm"]], bins=50, label=["Comprimento do bico","Profundidade do bico"])
plt.legend(loc="upper right")
plt.show()
plt.figure(figsize=(8,5))
plt.title("Massa Corporal Penguins Gentoo")
plt.hist([gento_df["body_mass_g"]], bins=50, label=["Massa Corporal"],color="blue")
plt.legend(loc="upper right")
plt.show()
```

```python

plt.figure(figsize=(8,5))
plt.title("Caracteristicas Penguins Chinstrap")
plt.hist([chinstrap_df["bill_length_mm"],chinstrap_df["bill_length_mm"]], bins=50, label=["Comprimento do bico","Profundidade do bico"])
plt.legend(loc="upper right")
plt.show()
plt.figure(figsize=(8,5))
plt.title("Massa Corporal Penguins Chinstrap")
plt.hist([chinstrap_df["body_mass_g"]], bins=50, label=["Massa Corporal"],color="blue")
plt.legend(loc="upper right")
plt.show()
```
