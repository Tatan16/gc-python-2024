### Discretizar atributos numericos que possui os valores abaixo em 3 intervalos

### 0,1,2,6,6,9,10,10,10,13,16,17,18,20,20,21,21,25

### Larguras iguais
### Frequencias iguais 
### Inspeçao visual

```python
import numpy as np
import pandas as pd
import cv2

### Larguras iguais
list1 = [0,1,2,6,6,9,10,10,10,13,16,17,18,20,20,21,21,25]
labels = ["baixo", "medio", "alto"]
# bins = np.linspace(min(list), max(list),4)
list_discretizada = pd.cut(list1, 3, labels=labels)
print(list_discretizada)
```

```python
# Frequencias Iguais
list2 = [0,1,2,6,6,9,10,10,10,13,16,17,18,20,20,21,21,25]
df = pd.DataFrame({"Valor":list2})

num_intervalos = 3

df["Intervalo"] = pd.qcut(df["Valor"], q=num_intervalos, labels=["Baixo", "Medio", "Alto"])
df
```

```python
# Carregando a imagem
img = cv2.imread('imagem.jpg', cv2.IMREAD_GRAYSCALE)

# Convertendo a imagem para um array 1D
pixels = img.flatten()

# Criando um DataFrame com os valores dos pixels
df_2 = pd.DataFrame({'pixel': pixels})

# Discretizando os pixels em 3 categorias
df_2['intensidade'] = pd.qcut(df['pixel'], q=3, labels=['escuro', 'médio', 'claro'])

# Convertendo o DataFrame de volta para uma imagem
img_discretizada = df_2['intensidade'].to_numpy().reshape(img.shape)

# Visualizando a imagem discretizada
cv2.imshow('Imagem Discretizada', img_discretizada)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
