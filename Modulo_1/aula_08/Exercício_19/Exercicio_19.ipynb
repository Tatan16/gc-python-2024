```python
op = penguinsExtended_df.groupby(["species"])[['bill_length_mm', 'flipper_length_mm', 'body_mass_g']].mean()
```

```python
op
```

```python
plt.boxplot(op['bill_length_mm'])
plt.title("Boxplot das bill length")
plt.ylabel(op['bill_length_mm'])
plt.show()
```

```python
plt.boxplot(op["flipper_length_mm"])
plt.title("Boxplot do Flipper Length")
plt.ylabel(op["flipper_length_mm"])
plt.show()
```

```python
plt.boxplot(op['body_mass_g'])
plt.title("Boxplot do Body Mass")
plt.ylabel(op['body_mass_g'])
plt.show()
```
