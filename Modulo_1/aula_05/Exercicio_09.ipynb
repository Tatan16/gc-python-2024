```python
import pandas as pd
```

```python
# Trazendo os DataFrames da outra pasta
dataframe1 = pd.read_json("../aula_03/meu_dataframe_animes.json")
dataframe2 = pd.read_json("../aula_03/meu_dataframe_desenhos.json")
dataframe3 = pd.read_json("../aula_03/meu_dataframe_filmes.json")

# Criando novos DataFrames com novas linhas
df1_novasLinhas = pd.DataFrame({

    "Meus_animes_favoritos":["One punch man", "Bleach", "Seven deadly sins"],
    "Ano_de_Lançamento":[2009, 2001,2012],
    "Nota_de_quanto_gosto":[9,9,8]
})

df2_novasLinhas = pd.DataFrame({

    "Meus_desenhos_favoritos":["Barney", "Bakugan", "Clarence"],
    "An_de_lançamento":[1992,2007,2014],
    "Nota_de_quanto_gosto":[6, 7, 7]
})

df3_novasLinhas = pd.DataFrame({

    "Meus_Filmes_Favoritos":["O relampago mcqueen", "As cronicas de narnia", "Avatar"],
    "Ano_de_lançamento":[2006,2005,2009],
    "Nota_de_quanto_gosto":[8,10,9]
})

# Concatenando o antigo com o novo dataframe
df1 = pd.concat([dataframe1, df1_novasLinhas], ignore_index=True, axis=0)
df2 = pd.concat([dataframe2, df2_novasLinhas], ignore_index=True, axis=0)
df3 = pd.concat([dataframe3, df3_novasLinhas], ignore_index=True, axis=0)

# Renomeando as columnas do DataFrame
df1.rename(columns={"Meus_animes_favoritos":"Meus_favoritos", "Ano_de_Lançamento":"Lançamento", "Nota_de_quanto_gosto":"Valoração"}, inplace=True)
df2.rename(columns={"Meus_desenhos_favoritos":"Meus_favoritos", "An_de_lançamento":"Lançamento", "Nota_de_quanto_gosto":"Valoração"}, inplace=True)
df3.rename(columns={"Meus_Filmes_Favoritos":"Meus_favoritos", "Ano_de_lançamento":"Lançamento", "Nota_de_quanto_gosto":"Valoração"}, inplace=True)

novo_dataframe = pd.concat([df1, df2, df3], ignore_index=True, axis=0)
```
